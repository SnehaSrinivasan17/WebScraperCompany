*   Make a request to the website
*   Put HTML in BeautifulSoup
*   Parse the item and use it with Pandas





```python
import pandas as pd
from bs4 import BeautifulSoup #extract data from HTML sites
import requests #make requests to the web pages
```

# Making request to the webpage

```python
r = requests.get('https://www.moderndatastack.xyz/companies')
```

```python
r
```

```python
r.status_code #sucessfully made the request
```

```python
r.text #html of the website
```

```python
type(r.text) #hard to parse data in str format
```

#  Beautiful Soup

```python
soup = BeautifulSoup(r.text,"html.parser") #creates a soup object which can be parsed
```

```python
soup
```

```python
type(soup)
```

```python
soup.findAll("a", {"class" : "mds-text-primary text-nowrap text-decoration-underline mds-fw-md" }) #finds all the URLs which are in the specified class
```

```python
type(soup.findAll("a", {"class" : "mds-text-primary text-nowrap text-decoration-underline mds-fw-md" }))
```

```python
len(soup.findAll("a", {"class" : "mds-text-primary text-nowrap text-decoration-underline mds-fw-md" })) # number of URLs in the page
```

```python
companies = soup.findAll("a", {"class" : "mds-text-primary text-nowrap text-decoration-underline mds-fw-md" })
```

```python
for company in companies:
  print(company)
```

```python
print(company)
```

```python
type(company)
```

```python
company.text
```

```python
company['href']
```

```python
data = []
for company in companies:
  company_name = company.text
  company_name = company_name[:int(len(company_name)/2)]
  print(company_name,company['href'])
  data.append({'company_name':company_name,'company_url':company['href']})
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.to_csv('output.csv')
```
